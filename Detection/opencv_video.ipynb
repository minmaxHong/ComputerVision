#### 경로 지정하기

```python
import cv2

video_input_path = "/content/ComputerVision/data/Image/Night_Day_Chase.mp4"
video_output_path = "/content/ComputerVision/data/Image/Night_Day_Chase_output.mp4"

cap = cv2.VideoCapture(video_input_path)
codec = cv2.VideoWriter_fourcc(*'XVID')

# == video size ==
vid_size = (round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))

# == video fps ==
vid_fps = cap.get(cv2.CAP_PROP_FPS)
total_fps = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))  # 총 fps로 반환은 float

# == videowritter ==
vid_writer = cv2.VideoWriter(video_output_path, codec, vid_fps, vid_size)
```

#### frame별로 읽어들이기

```python
import time

start = time.time()
index = 0

while True:
  hasFrame, img_frame = cap.read()
  if not hasFrame:
    print("Can not search file anymore")
    break

  index += 1
  print("frame : ", index , "처리 완료")

  cv2.rectangle(img_frame, (300, 100), (800, 400), color = (0, 255, 0), thickness = 2)
  text = "frame : {}".format(index)
  cv2.putText(img_frame, text, (300, 95), cv2.FONT_HERSHEY_SIMPLEX, 0.7, color = (255, 0, 0), thickness = 2)

  vid_writer.write(img_frame)

print("완료 시간 : {}".format(time.time() - start))
vid_writer.release()
cap.release()
```
